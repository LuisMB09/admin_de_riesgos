```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def collar_strategy(stock_price, strike_put, put_premium, strike_call, call_premium, stock_entry_price):
    stock_prices = np.linspace(stock_price * 0.5, stock_price * 1.5, 100)
    
    # Largo en el activo
    stock_pnl = stock_prices - stock_entry_price
    
    # Put de proteccin (long put)
    put_pnl = np.maximum(strike_put - stock_prices, 0) - put_premium
    
    # Call cubierto (short call)
    call_pnl = -(np.maximum(stock_prices - strike_call, 0)) + call_premium
    
    # PnL total de la estrategia
    total_pnl = stock_pnl + put_pnl + call_pnl
    
    plt.figure(figsize=(10, 6))
    plt.plot(stock_prices, stock_pnl, '--', label='Stock PnL', alpha=0.7)
    plt.plot(stock_prices, put_pnl, '--', label='Put PnL', alpha=0.7)
    plt.plot(stock_prices, call_pnl, '--', label='Call PnL', alpha=0.7)
    plt.plot(stock_prices, total_pnl, label='Total Strategy PnL', linewidth=2)
    
    plt.axhline(0, color='black', linewidth=1)
    plt.xlabel("Stock Price at Expiration")
    plt.ylabel("Profit / Loss")
    plt.legend()
    plt.title("Collar Strategy PnL")
    plt.grid()
    plt.show()

# Parmetros de la estrategia
stock_price = 100   # Precio actual del activo
stock_entry_price = 100  # Precio de entrada en la compra
strike_put = 90  # Strike del put comprado
put_premium = 3  # Costo del put
strike_call = 110  # Strike del call vendido
call_premium = 2  # Prima recibida por el call

collar_strategy(stock_price, strike_put, put_premium, strike_call, call_premium, stock_entry_price)
```
