```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.facecolor'] = 'lightgrey'
```

## Procter & Gamble

```python
ist_pg = pd.read_excel('PG/ist.xlsx', sheet_name=1)
ist_pg = ist_pg.set_index('Unnamed: 0')
ist_pg.index.name = None

bs_pg = pd.read_excel('PG/bs.xlsx', sheet_name=1)
bs_pg = bs_pg.set_index('Unnamed: 0')
bs_pg.index.name = None

cf_pg = pd.read_excel('PG/cf.xlsx', sheet_name=1)
cf_pg = cf_pg.set_index('Unnamed: 0')
cf_pg.index.name = None
```

```python
current_ratio = bs_pg.loc['Total Current Assets'] / bs_pg.loc['Total Current Liabilities']
solvency = bs_pg.loc['Total Assets'] / bs_pg.loc['Total Liabilities']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(solvency, label='Solvency Ratio', color='navy')
ax1.set_title('Solvency Ratio over Time')
ax1.set_ylabel('Solvency Ratio')
ax1.legend()

ax2.plot(current_ratio, label='Current Ratio', color='navy')
ax2.set_title('Current Ratio over Time')
ax2.set_ylabel('Current Ratio')
ax2.legend()

plt.tight_layout()
```
