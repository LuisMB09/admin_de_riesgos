```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.facecolor'] = 'lightgrey'
```

```python
ist = pd.read_excel('PG/ist.xlsx', sheet_name=1)
ist = ist.set_index('Unnamed: 0')
ist.index.name = None

bs = pd.read_excel('PG/bs.xlsx', sheet_name=1)
bs = bs.set_index('Unnamed: 0')
bs.index.name = None

cf = pd.read_excel('PG/cf.xlsx', sheet_name=1)
cf = cf.set_index('Unnamed: 0')
cf.index.name = None
```

```python
current_ratio = bs.loc['Total Current Assets'] / bs.loc['Total Current Liabilities']
solvency = bs.loc['Total Assets'] / bs.loc['Total Liabilities']


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.plot(solvency, label='Solvency Ratio', color='blue')
ax1.set_title('Solvency Ratio over Time')
ax1.set_ylabel('Solvency Ratio')
ax1.legend()

ax2.plot(current_ratio, label='Current Ratio', color='orange')
ax2.set_title('Current Ratio over Time')
ax2.set_ylabel('Current Ratio')
ax2.legend()

plt.tight_layout()
```
